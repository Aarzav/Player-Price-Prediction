# Satisfying the assumption of Machine Learning Regression

## 1. Linearity

```python
import statsmodels.api as sm

# Fit the multiple linear regression model
model = sm.OLS(y,x).fit()

# Print the summary of the regression analysis
print(model.summary())
```

Durbin-Watson: is close to 2 which is 1.934 so that the error have independency 

Adj. R-squared: 97% of data explain by x so that means 3% error is present

model significant: F-statistic is 631.2 which state that model is significant from this we can say it has linearity   

## 2. residual normality

```python
model_resid = model.resid 
probplot = sm.ProbPlot( model_resid )
plt.figure( figsize = (8, 6) )
probplot.ppplot( line='45' )
plt.title( "Normal P-P Plot of Regression Standardized Residuals" )
plt.show()
```

In a P-P plot, the observed data (in this case, the residuals) are plotted against the expected values from a theoretical distribution, typically a normal distribution. It helps you assess whether the residuals follow a specific distribution, such as a normal distribution.

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(model_resid, shade=True)
plt.title("Kernel Density Estimation (KDE) Plot of Regression Residuals")
plt.xlabel("Residuals")
plt.ylabel("Density")
plt.xlim(-10, 10)
plt.show()
```

 it provides a visual tool to assess the distribution of residuals. When assessing normality, look for signs of symmetry, bell-shaped curves, and smooth tails in the KDE plot. Deviations from these characteristics may suggest departures from normality in the residuals, which could impact the validity of assumptions in regression analysis.

## 3. Independence of Errors

```python
model_pred = model.predict(x)
plt.scatter(model_pred, model_resid)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='red', linestyle='--')  # Add a horizontal line at y=0 for reference
plt.title('Independence of Errors')
plt.show()
```

the independence of errors in our regression model suggests that the residuals exhibit no discernible patterns or correlations with either the predictor variables or their lagged values. Both visual examination and formal statistical tests support the validity of the independence of errors assumption. 

## 4. Homoscedasticity

```python
# Sample code to check homoscedasticity assumption
plt.scatter(model_pred,model_resid)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='red', linestyle='--')
plt.axhline(y=3*model_resid.std(), color='green', linestyle='--')  # Check for constant spread
plt.axhline(y=-3*model_resid.std(), color='green', linestyle='--')
plt.show()
```

 the assessment of homoscedasticity, or the constant variance of residuals, indicates that the residuals exhibit a consistent spread around some across the range of predictor variables. 

## 5. Auto-Correlation

```python
sm.graphics.tsa.plot_acf(model_resid, lags=40)  # 'lags' specifies the number of lags to show
plt.title("Autocorrelation Function (ACF) Plot for Residuals")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.show()
```

To determine whether a regression model has autocorrelation (i.e., whether the residuals exhibit a pattern of correlation with their lagged values), you can create an autocorrelation plot (ACF plot) for the residuals. If the ACF plot shows significant correlations at certain lag values

So it has negative auto-correlation because it is scattered and distributed nearer to the x axis

In a dataset comprising 962 columns We removed the MultiCollinearity to 10 Columns, a model may achieve an impressive 99.1% accuracy, but the practicality of applying this model becomes challenging when users must input all 962 features.

Feature selection becomes paramount in such scenarios, allowing us to identify and retain only the most informative attributes while discarding less relevant ones. By reducing the dimensionality of the dataset through feature selection, we streamline the user experience, simplify data collection, and potentially improve model interpretability, all without sacrificing significant predictive power.

This approach not only makes the model more accessible but also optimizes resource utilization and ensures that only essential information is considered for decision-making.

So, We are going to reduce 962 columns to 20 columns according to value_eur

```python
from sklearn.feature_selection import SelectKBest, f_regression

selector = SelectKBest(score_func=f_regression, k=10)  

x_new = selector.fit_transform(x, y)


selected_indices = selector.get_support(indices=True)

selected_feature_names = x.columns[selected_indices]

x_new_df = pd.DataFrame(x_new, columns=selected_feature_names)
```

```python
NN = MLPRegressor (max_iter =300, activation ='relu' , hidden_layer_sizes = (100,100))
NN.fit(x_train_new_pca, y_train)
y_pred = NN.predict(x_test_new_pca)
print('Training Score:',NN.score(x_train_new_pca,y_train))
print('Testing Score:',NN.score(x_test_new_pca,y_test))
```

```python
algorithm_name='MLPRegressor With PCA and Feature Selection'
mse = mean_squared_error(y_pred,y_test)
print('Mean Squared Error:', mse)
rmse = np.sqrt(mse)
print('Root Mean Squared Error:', rmse)
mae = mean_absolute_error(y_pred,y_test)
print('Mean Absolute Error:', mae)
r2 = r2_score(y_pred,y_test)
print('R-squared:', r2)
```
